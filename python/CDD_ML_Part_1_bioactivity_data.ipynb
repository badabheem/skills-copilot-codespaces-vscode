## Build gnn model

### Subtask:
Design and implement a GNN model architecture suitable for molecular graph analysis and oral bioavailability prediction.


**Reasoning**:
To build a GNN model, I will use PyTorch Geometric. First, I need to install the necessary libraries, including `torch` and `torch_geometric`. I will then define a simple GNN architecture using GCN layers, which are suitable for this kind of graph classification task. The model will take graph-level features as input and produce a class prediction as output.



```
!pip install torch
!pip install torch_geometric
import torch
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv, global_mean_pool
import torch.nn.functional as F

def mol_to_graph_data(mol):
    # Atom features
    atom_features = []
    for atom in mol.GetAtoms():
        atom_features.append([
            atom.GetAtomicNum(),
            atom.GetDegree(),
            atom.GetFormalCharge(),
            atom.GetNumRadicalElectrons(),
            atom.GetHybridization(),
            atom.GetIsAromatic(),
            atom.IsInRing(),
        ])
    x = torch.tensor(atom_features, dtype=torch.float)

    # Edge features
    edge_indices = []
    for bond in mol.GetBonds():
        edge_indices.append([bond.GetBeginAtomIdx(), bond.GetEndAtomIdx()])
    edge_index = torch.tensor(edge_indices, dtype=torch.long).t().contiguous()

    return Data(x=x, edge_index=edge_index)

class GCN(torch.nn.Module):
    def __init__(self, num_node_features, num_classes):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(num_node_features, 16)
        self.conv2 = GCNConv(16, 32)
        self.fc = torch.nn.Linear(32, num_classes)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index

        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        x = F.relu(x)
        x = global_mean_pool(x, data.batch)
        x = self.fc(x)
        return F.log_softmax(x, dim=1)

# Determine the number of node features from the first graph
sample_graph_data = mol_to_graph_data(molecular_graphs[0])
num_node_features = sample_graph_data.num_node_features
num_classes = len(label_map)

# Instantiate the model
model = GCN(num_node_features=num_node_features, num_classes=num_classes)
print(model)
```

## Deploy model

### Subtask:
Prepare the trained model for deployment, which might involve saving the model and creating an interface for making predictions on new molecular data.


**Reasoning**:
Save the trained model's state dictionary to a file and define the prediction function as instructed.



```
import torch
from torch_geometric.data import Data
from rdkit import Chem

# 1. Save the trained GNN model's state dictionary
torch.save(model.state_dict(), 'gnn_model_state_dict.pth')
print("Trained model state dictionary saved to gnn_model_state_dict.pth")

# 2. Define a prediction function
def predict_bioactivity(smiles_string, model_path='gnn_model_state_dict.pth', num_node_features=7, num_classes=3, label_map=None):
    """
    Predicts the bioactivity class for a given SMILES string using the trained GNN model.

    Args:
        smiles_string (str): The SMILES representation of the molecule.
        model_path (str): Path to the saved model state dictionary.
        num_node_features (int): Number of features per node in the molecular graph.
        num_classes (int): Number of output classes for prediction.
        label_map (dict): A dictionary mapping numerical labels back to bioactivity classes.

    Returns:
        str: The predicted bioactivity class ('active', 'intermediate', or 'inactive').
             Returns None if the SMILES string is invalid or prediction fails.
    """
    if label_map is None:
         # Assuming a default label map if not provided
         label_map = {0: "active", 1: "intermediate", 2: "inactive"}

    # Convert SMILES to RDKit molecule
    mol = Chem.MolFromSmiles(smiles_string)
    if mol is None:
        print(f"Invalid SMILES string: {smiles_string}")
        return None

    # Convert RDKit molecule to PyTorch Geometric Data object (without label)
    atom_features = []
    for atom in mol.GetAtoms():
        atom_features.append([
            atom.GetAtomicNum(),
            atom.GetDegree(),
            atom.GetFormalCharge(),
            atom.GetNumRadicalElectrons(),
            atom.GetHybridization(),
            atom.GetIsAromatic(),
            atom.IsInRing(),
        ])
    x = torch.tensor(atom_features, dtype=torch.float)

    edge_indices = []
    for bond in mol.GetBonds():
        edge_indices.append([bond.GetBeginAtomIdx(), bond.GetEndAtomIdx()])
    # Handle molecules with no bonds (e.g., single atom)
    if not edge_indices:
        edge_index = torch.empty((2, 0), dtype=torch.long)
    else:
        edge_index = torch.tensor(edge_indices, dtype=torch.long).t().contiguous()

    data = Data(x=x, edge_index=edge_index)

    # Load the model
    # Instantiate the model with the same architecture parameters
    model = GCN(num_node_features=num_node_features, num_classes=num_classes)
    model.load_state_dict(torch.load(model_path))
    model.eval() # Set model to evaluation mode

    # Add a batch dimension for prediction (PyTorch Geometric expects a batch)
    data.batch = torch.zeros(data.num_nodes, dtype=torch.long)

    # Perform prediction
    with torch.no_grad():
        out = model(data)
        # Get the predicted class index
        _, predicted_class_index = torch.max(out, 1)

    # Map the index back to the bioactivity class string
    predicted_class = label_map.get(predicted_class_index.item(), "Unknown")

    return predicted_class

# 3. Test the prediction function with a sample SMILES string
# Using a sample SMILES from the original df_preprocessed
sample_smiles = df_preprocessed['canonical_smiles'].iloc[0] # Get the first SMILES string
predicted_bioactivity = predict_bioactivity(sample_smiles, label_map={0: "active", 1: "intermediate", 2: "inactive"})
print(f"\nSample SMILES: {sample_smiles}")
print(f"Predicted Bioactivity Class: {predicted_bioactivity}")

# Test with another sample SMILES
sample_smiles_2 = df_preprocessed['canonical_smiles'].iloc[3] # Get the fourth SMILES string
predicted_bioactivity_2 = predict_bioactivity(sample_smiles_2, label_map={0: "active", 1: "intermediate", 2: "inactive"})
print(f"\nSample SMILES 2: {sample_smiles_2}")
print(f"Predicted Bioactivity Class: {predicted_bioactivity_2}")
```
